```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(10)
y1 = x**2
y2 = 2*x + 3
print(x)
print(y1)
print(y2)
```

```python
# Line Plot
plt.style.use('seaborn')
plt.plot(x, y1, color='red', label='Apple', marker='o')
plt.plot(x, y2, color='green', label='Kiwi', linestyle='dashed')
plt.xlabel('Time')
plt.ylabel('Price')
plt.title('Time vs Price')
plt.legend()
plt.show()
```

```python
# Scatter Plot
plt.scatter(x, y1)
plt.scatter(x, y2)
plt.show()
```

```python
# Bar Graph
plt.bar([1,4,7], [10,20,30], width=0.5, label="Current", tick_label=['Gold', 'Silver', 'Platinum'])
plt.bar([1.5, 4.5, 7.5], [5,10,10], width=0.5, label='Next')
plt.legend()
plt.show()
```

```python
# Pie Chart
subjects = "Maths", "Eng", "SST", "Science"
marks = [40, 60, 20, 70]
plt.pie(marks, labels=subjects, explode=(0,0,0.1,0))
plt.show()
```

```python
Xsn = np.random.randn(100)
u = 60
sigma = 8
Xsn = np.round(Xsn*sigma + u)
print(Xsn)
```

```python
# Histogram
plt.hist(Xsn)
plt.show()
```

```python
# Multivariate Normal Distribution
mean = np.array([0.0, 0.0])
cov = np.array([[1,0],[0,1]])

dist = np.random.multivariate_normal(mean, cov, 500)
print(dist.shape)

plt.scatter(dist[:,0], dist[:,1])
plt.show()
```
